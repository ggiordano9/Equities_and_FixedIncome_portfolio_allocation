```python
'''
Github link for a module made (yesg - p) to retreive ESG Scores from yahoo finance, including historical ones:
https://github.com/Lienus10/yesg/blob/master/yesg/main.py

'''

import yesg
import pandas as pd
import calendar
import urllib.request
import json
import requests

sp_500 = pd.read_csv('sandp500.csv') 

# Downloads historic ESG ratings and returns it as a dataframe

def get_historic_esg(ticker):
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) '
                             'Chrome/50.0.2661.102 Safari/537.36'}

    response = requests.get('https://query2.finance.yahoo.com/v1/finance/esgChart', params={"symbol": ticker},
                            headers=headers)

    try:
        df = pd.DataFrame(response.json()["esgChart"]["result"][0]["symbolSeries"])
        df["Date"] = pd.to_datetime(df["timestamp"], unit="s")

        df = df.rename(columns={"esgScore": "Total-Score", "environmentScore": "E-Score", "socialScore": "S-Score",
                                "governanceScore": "G-Score"})
        df["Ticker"] = ticker

        return df[['Date', 'Total-Score', 'E-Score', 'S-Score', 'G-Score', 'Ticker']].set_index('Date')

    except:
        print('An error has occurred. The ticker symbol might be wrong or you might need to wait to continue.')


'''
Or, using yesg module, could just do:
df = yesg.get_historic_esg('JPM')
df

'''
```

```python
appended_data = []
for Ticker in sp_500['Ticker']:
    a_data = get_historic_esg(Ticker)
    appended_data.append(a_data)
appended_data = pd.concat(appended_data)
appended_data


```
