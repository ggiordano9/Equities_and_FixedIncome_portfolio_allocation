```python
import pandas as pd
munis = pd.read_csv('municipal_bonds.csv')
treas = pd.read_csv('treasury_bonds.csv')
```

```python
treas
```

```python
'''
Use the tax-equivalent-yield formula to compare the real return on a muni bond with a corporate bond.
The formula is: Tax Equivalent Yield = Tax-Free Yield / (1 – Tax Rate).

'''
income_response = float(input("What is your annual income? (if married, please respond the joint annual income): "))
age_response = float(input("What is your age? "))

if income_response >= 0 and income_response <= 10275:
    tax_rate = .10
elif  income_response >=10275 and  income_response < 41775:
    tax_rate = .12
elif  income_response >=41775 and  income_response <89075:
    tax_rate = .22
elif  income_response >=89075 and  income_response <170050:
    tax_rate = .24
elif  income_response >=170050 and  income_response < 215950:
    tax_rate = .32
elif  income_response >=215950 and  income_response <539900:
    tax_rate = .35
elif  income_response >=539900:
    tax_rate = .37

print(tax_rate)
```

```python
treas_tax_equivalet_yield =( munis['Coupon']) / (1-tax_rate)
```

```python
treas_tax_equivalet_yield
```

```python
munis_vs_treas = pd.DataFrame(treas_tax_equivalet_yield)
```

```python
munis_vs_treas = pd.DataFrame(munis[['Issuer Name', 'CUSIP', 'Coupon']])
```

```python
munis_vs_treas['treasury_equivalet_yield'] = treas_tax_equivalet_yield
```

```python
munis_vs_treas
```

```python
treas_weight = 0
muni_weight = 0
if tax_rate == .10 or tax_rate == .12:
    treas_weight = .100
    muni_weight =0
if tax_rate == .22 or tax_rate == .24 or tax_rate == .32:
    treas_weight = .50
    muni_weight = .50
if tax_rate == .35 or tax_rate == .37:
    treas_weight = 0
    muni_weight = .100
print("the portfolio weight for treasury bonds should be", treas_weight * 100, "% since your income tax rate is ", tax_rate)
print("the portfolio weight for municipal bonds should be", muni_weight * 100, "% since your income tax rate is ", tax_rate)
```

```python
import numpy as np
equity_allocation = 100 - age_response
fixed_allocation = 100 - equity_allocation
print("equity allocation", equity_allocation)
print("fixed income allocation", fixed_allocation)
treas_yield_chosen = np.mean(munis_vs_treas['treasury_equivalet_yield'])
muni_yield_chosen = np.mean(munis_vs_treas['Coupon'])
fixed_allocation_return = (treas_weight * treas_yield_chosen) + (muni_weight * muni_yield_chosen)
print("fixed_return", fixed_allocation_return)
equity_annual_return = .27
```

```python
portfolio_annual_return = (equity_allocation / 100) * equity_annual_return + (fixed_allocation/100) * ((fixed_allocation_return * 2)/100)
print("Your annual portfolio return is", portfolio_annual_return)
```
